### Calculadora Simples em Python

Esta seção apresenta uma calculadora básica que realiza operações de adição, subtração, multiplicação e divisão. Cada função é definida para uma operação específica e um bloco principal demonstra como usá-las, permitindo a interação do usuário.

```python
# Função para realizar a adição de dois números
def adicao(x, y):
    return x + y

# Função para realizar a subtração de dois números
def subtracao(x, y):
    return x - y

# Função para realizar a multiplicação de dois números
def multiplicacao(x, y):
    return x * y

# Função para realizar a divisão de dois números
def divisao(x, y):
    # Verifica se o divisor é zero para evitar erro de divisão por zero
    if y == 0:
        return "Erro! Divisão por zero não é permitida."
    else:
        return x / y

# Exibe o menu de operações para o usuário
print("Selecione a operação:")
print("1. Adição")
print("2. Subtração")
print("3. Multiplicação")
print("4. Divisão")

# Loop principal para permitir que o usuário faça várias operações
while True:
    # Pede ao usuário para escolher uma opção
    escolha = input("Digite sua escolha (1/2/3/4): ")

    # Verifica se a escolha é uma das opções válidas
    if escolha in ('1', '2', '3', '4'):
        try:
            # Solicita o primeiro número ao usuário e o converte para float
            num1 = float(input("Digite o primeiro número: "))
            # Solicita o segundo número ao usuário e o converte para float
            num2 = float(input("Digite o segundo número: "))
        except ValueError:
            # Captura erro se o usuário digitar algo que não é um número
            print("Entrada inválida. Por favor, digite um número.")
            continue

        # Realiza a operação com base na escolha do usuário
        if escolha == '1':
            print(num1, "+", num2, "=", adicao(num1, num2))

        elif escolha == '2':
            print(num1, "-", num2, "=", subtracao(num1, num2))

        elif escolha == '3':
            print(num1, "*", num2, "=", multiplicacao(num1, num2))

        elif escolha == '4':
            print(num1, "/", num2, "=", divisao(num1, num2))

        # Pergunta ao usuário se ele deseja fazer outra operação
        proxima_operacao = input("Deseja realizar outra operação? (sim/não): ")
        if proxima_operacao.lower() != 'sim':
            break
    else:
        # Mensagem de erro se a escolha da operação for inválida
        print("Entrada inválida. Por favor, digite 1, 2, 3 ou 4.")
```
